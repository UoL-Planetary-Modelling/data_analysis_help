    - array([cftime.DatetimeNoLeap(year, month, day, hour, minute, second, microsecond, has_year_zero=True)],  dtype=object)
    
    - Let's break this down:
       
        - array() 
            - (TODO - confirm) This represents an xarray DataArray data structure. Similar to numpy arrays, DataArrays can hold multiple values of the same data type or of different data types. Hence the first element is a list, and the second element is an argument specifying the data type of the array elements.
        - dtype=object
            - This means that the array can hold elements of any data type, as indicated by the "object" data type.
        - cftime.DatetimeNoLeap()
            - This is a specific type of date and time representation provided by the cftime module. It's used to represent dates and times with a specific calendar and handling of leap years. cftime.DatetimeNoLeap() is a class representing a date and time instance without leap years, which means it's suitable for calendars without leap years, such as the "No Leap" calendar.
        - has_year_zero=True 
            - This is an argument passed to the DatetimeNoLeap constructor. It specifies whether the calendar system includes a year zero or not. In this case, has_year_zero=True indicates that the calendar system includes a year zero.

# Common problems and solutions with examples

## Converting from YYYYMM (or equivalent) date format to cftime.DatetimeNoLeap object. Example shows climatological calculations from data that originally has this format.

- MMK: I encountered this problem when working with MIPAS data which had a time coordinate with this format. The dataset consisted of 4 years of monthly means, and I wanted to calculate the climatology by grouping the data and averaging using ds.groupby("time.month").mean(dim="time") methods. 

- The "time.month" accessor will not work on dates with format YYYYMM

- One way you can solve this problem is by converting dates with this format, or similar formats (e.g. YYYY-MM-DD) to cftime.DatetimeNoLeap objects, the below snippets shows you how.

- NOTE - computing e.g. ds_grouped = ds.groupby("time.month").mean(dim="time") shifts the data cyclicly by 1 month "to the right". E.g. for a full year of monthly means, indexing ds_grouped.isel(month=1) should correspond to Feb, but instead corresponds to Jan, similarly ds.isel(month=0) should be Jan, but instead is Dec. One way of addressing it is shown in the final snippet.

- Example shown is for monthly mean T data for the years 2008-2011 from the MIPAS satellite, zonally averaged at 40N.

---

To create a cftime.DatetimeNoLeap object, you can use

```python
import cftime

Y = 2010
M = 1
D = 2
H = 3
m = 4
s = 5

datetime_no_leap = cftime.DatetimeNoLeap(Y, M, D, H, m, s, has_year_zero=True)

datetime_no_leap
```

To access all of the individual components you can use the accessors: year, month, day, hour, minute and second.


```python

print(datetime_no_leap.year)
print(datetime_no_leap.month)
print(datetime_no_leap.day)
print(datetime_no_leap.hour)
print(datetime_no_leap.minute)
print(datetime_no_leap.second)
```

Below shows this in use on the example MIPAS data. Note the adherence to the date convention used by xarray to handle dates - i.e. the date for average quantities corresponds to the end of the time bound of the averaging period. E.g. if you have an average for Dec 2010, the date shown will be 2011-01-01... 

```python
%%time
import xarray as xr
import cftime
# Read in data for example
path = "./example_data/"
file = "MIPAS_T_40N_ZM_2008_2011.nc"
ds = xr.open_dataset(path + file)
# Inspect
ds
```

```python
# Loop over time values and convert to cftime.DatetimeNoLeap object
cftime_values = []
for time_str in ds['time'].values:
    
    year = int(str(time_str)[0:4]) # Extract year 
    month = int(str(time_str)[4:])+1 # Extract month and add 1 due to xarray convention
    if month == 13:
        year = year + 1
        month = 1
    
    # Create a cftime.DatetimeNoLeap object
    datetime_no_leap = cftime.DatetimeNoLeap(year, month, 1, 0, 0, 0, has_year_zero=True)
    
    # Append the cftime.DatetimeNoLeap object to the list
    cftime_values.append(datetime_no_leap)
    
# Assign cftime_values list to time dimension in ds
ds['time'] = xr.DataArray(cftime_values, dims="time")
# Inspect dataset
ds
```

```python
# Perform grouping and plot

# Note - this grouping produces a "time" dimensions called "month" with values from 1 <= month <= 12.
# This can be confusing as accessing the data entries using isel is done using the 
# positional index, e.g. month 1 is accessed as ds.isel(month=0)
# Accessing the first data entry using "sel" is done by the dimension value ds.sel(month=1)

# This is made further confusing because the grouping operation shifts the data to the right 
# by one month, so ds.isel(month=0) = ds.sel(month=1) actually gives December data.

# One way of overcoming this is using the .shift() operation

# This operation groups the data and changes the order from Dec, Jan, Feb, ..., Nov
# to Jan, Feb, ..., Nov.
ds_clim = ds.groupby("time.month").mean(dim="time").shift(month=-1) 

# However, data for december is no longer here, so you can assign this separately using .loc
# Note that .loc doesn't work with positional indices, but instead with the dimension value,
# hence why we use .loc[dict(month=12)] - see https://docs.xarray.dev/en/latest/user-guide/indexing.html#quick-overview
# for more details
ds_clim.loc[dict(month=12)] = ds.groupby("time.month").mean(dim="time").isel(month=0)

# You now have monthly climatologies, and the correct months at the correct positions in the array
ds_clim['T'].plot(y="z")
```
